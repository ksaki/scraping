```python
#scraping wto cases
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
import pandas as pd
import csv
import numpy as np
import chardet
import re


#Phantom JS is no longer availabe. Use headless Chrome. 
options = Options()
options.binary_location = '/Applications/Google Chrome Canary.app/Contents/MacOS/Google Chrome Canary'
options.add_argument('--headless')

#access to case list page
driver = webdriver.Chrome(executable_path = "/Users/hiwatari/Desktop/KOYAMA RA Work/chromedriver", chrome_options = options)
driver.get("https://www.wto.org/english/tratop_e/dispu_e/dispu_status_e.htm")

#get case links
case_links = []
for a in driver.find_elements_by_xpath("//a"):
    temp = a.get_attribute('href')
    if temp != None:
        if '/english/tratop_e/dispu_e/cases_e' in temp:
            case_links.append(temp)

#access each case link, get current status and summary table
d = {"case":[],"status":[],"last_date":[],"short_title":[], "complainant":[], "respondent":[], "third_parties":[],
    "agreement_cited":[], "request_received":[], "url":[]}

for i, case_link in enumerate(case_links):
    # storing the url for this case to the data.
    d["url"].append(case_link)
    
    # accessing
    driver.get(case_link)
    time.sleep(1)
    
    # getting and storing a title of the case
    ds_title = driver.find_element_by_tag_name("h1")
    d["case"].append(ds_title.text)
    time.sleep(0.5)
    #get current status
    status_we = driver.find_element_by_xpath("//span[@class='paraboldcolourtext']")
    d["status"].append(status_we.text)
    # get when the "current" means
    status_current_time = driver.find_element_by_tag_name("b")
    d["last_date"].append(status_current_time.text)
    time.sleep(0.5)
    # get factor table
    factor_table = driver.find_elements_by_tag_name("td")
    # short title
    d["short_title"].append(factor_table[1].text)
    # complainant
    d["complainant"].append(factor_table[3].text)
    # respondent etc...
    time.sleep(0.5)
    d["respondent"].append(factor_table[5].text)
    d["third_parties"].append(factor_table[7].text)
    time.sleep(0.5)
    d["agreement_cited"].append(factor_table[9].text)
    d["request_received"].append(factor_table[11].text)
    print(i)
    
    time.sleep(3)

 #interrupted at i == 370 due to anomalous style of the page.
 #Continue the process with skipping that page in the next cell.
```
